```python
import geopandas as gpd
```

```python
import pandas as pd
states = gpd.read_file('/content/drive/MyDrive/India States/Indian_states.shp')
stcp = pd.read_csv("/content/drive/MyDrive/state_capital_population.csv")
states = states.merge(stcp, left_on='st_nm', right_on='State')
states
```

```python
import plotly.express as px
import matplotlib.pyplot as plt

ax = states.plot(figsize=(12, 12), edgecolor='black', alpha=0.5, color='white', legend=True)

for x, y, pop, label in zip(states.geometry.centroid.x, states.geometry.centroid.y, states['Population'], states["Capital"]):
    ax.scatter(x, y, s=pop/1000, alpha=0.3, cmap='viridis', linewidth=0.5, edgecolor='0.2')
    ax.annotate(label, (x, y), textcoords="offset points", xytext=(0,10), ha='center', fontsize='12' )

ax.axis('off')
plt.show()
```

Iterates over each state in the india dataframe, extracting the x and y coordinates of the state's centroid, the population of the state, and the name of the state capital and creates a scatter plot of the state capital's location on the map, with the size of the marker proportional to the population of the state, and a low transparency. 

```python
states["geometry"] = states["geometry"].centroid
INDPS = states.to_crs("EPSG:4326")

fig = px.scatter_mapbox(states, lat=INDPS.geometry.y, lon=INDPS.geometry.x, size='Population', color='Population', text='State', hover_name='Capital', size_max=50, mapbox_style="carto-positron", zoom=3, opacity=0.5, color_discrete_sequence=px.colors.qualitative.T10)

fig.update_layout(hovermode="closest",)

fig.show()
```

## 2. Connection graph of Go Air domestic flights of India from the database

```python
flightschedule = pd.read_excel("/content/drive/MyDrive/Flightschedule.xlsx")
# Considering only the Non-Stop routes
flightschedule = flightschedule[flightschedule["Routing"]=="Non Stop"]
flightschedule = flightschedule.reset_index()
flightschedule.head()
```

```python
from geopy.geocoders import Nominatim

counter=0

departures = set(flightschedule["Origin"].unique())
destinations = set(flightschedule["Destination"].unique())
locations = list(set(list(departures) + list(destinations)))
all = {}

G = Nominatim(user_agent="geoapiExercises")

for l in locations:
    all[l] = G.geocode(l)

all_df = pd.DataFrame(columns=['City', 'Latitude', 'Longitude'])


for key in all.keys():
    all_df.loc[counter] = {'City': key, 'Latitude': all[key].latitude, 'Longitude': all[key].longitude}
    counter += 1
```

```python
cumi_data = pd.DataFrame(columns=['Origin', 'Origin_Latitude', 'Origin_Longitude', 'Destination', 'Destination_Latitude', 'Destination_Longitude'])

for i in range(len(flightschedule)):
    cumi_data.loc[i] = {'Origin': flightschedule.iloc[i,:]['Origin'], 'Destination': flightschedule.iloc[i,:]['Destination'], 'Origin_Latitude': float(all_df[all_df['City']==flightschedule.iloc[i,:]['Origin']]['Latitude']), 'Origin_Longitude': float(all_df[all_df['City']==flightschedule.iloc[i,:]['Origin']]['Longitude']), 'Destination_Latitude': float(all_df[all_df['City']==flightschedule.iloc[i,:]['Destination']]['Latitude']), 'Destination_Longitude': float(all_df[all_df['City']==flightschedule.iloc[i,:]['Destination']]['Longitude']),}

cumi_data.head()
```

```python
paths = [dict(type='scattermapbox', mode='lines', hovertext = row['Destination'], line=dict(width=0.75, color='coral'),lon=[row['Origin_Longitude'], row['Destination_Longitude']], lat=[row['Origin_Latitude'], row['Destination_Latitude']], showlegend=True) for _, row in cumi_data.iterrows()]

fig = px.scatter_mapbox(all_df, lat='Latitude', lon='Longitude',color_discrete_sequence=px.colors.qualitative.T10, mapbox_style="carto-positron", center={"lat": 20, "lon": 78}, size_max=7, zoom=3,  color='City', text='City')
fig.update_traces(marker={'size': 10})

for p in paths:
    fig.add_trace(p)

fig.show()
```

First defines a list of traces called paths by iterating over the rows of a Pandas DataFrame called cumi_data. Each trace in paths is a scattermapbox trace, which is a type of trace that shows a set of points and lines on a Mapbox map. Each trace in paths connects two locations (origin and destination) with a colored line and displays the destination city as the hover text.
